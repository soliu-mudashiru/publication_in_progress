```python
# plotting the roc curve for IDOCRIW-COCOSO GPOT

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv("gpi_storativity_data.csv")

# Rename column if needed
if "GPI" in df.columns:
    df.rename(columns={"GPI": "GPI_Index"}, inplace=True)

# Apply K-means to storativity (NOT GPI)
storativity_values = df["Storativity"].values.reshape(-1, 1)
kmeans = KMeans(n_clusters=2, random_state=42).fit(storativity_values)

# Assign binary labels based on cluster (0 = low, 1 = high)
cluster_means = df.groupby(kmeans.labels_)["Storativity"].mean()
high_cluster = cluster_means.idxmax()
df["Storativity_Class"] = [1 if label == high_cluster else 0 for label in kmeans.labels_]

# Use GPI as predictor
fpr, tpr, thresholds = roc_curve(df["Storativity_Class"], df["GPI_Index"])
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (IDOCRIW-COCOSO GPI predicting K-means-Storativity class)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("IDOCRIW-COCOSO roc_storativity_kmeans.tif", dpi=600)
plt.show()
```

```python
import pandas as pd
from scipy.stats import wilcoxon
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV
df = pd.read_csv("winconxon_test_data.csv")  # Change to your file path

# Define models and true values
models = ["IDOCRIW_COCOSO", "CRITIC_COCOSO", "MeanWeighted", "MeanWeighted_COCOSO"]
y_true = df["Storativity"]

# Calculate absolute errors
errors = {model: np.abs(df[model] - y_true) for model in models}

# Wilcoxon signed-rank tests: IDOCRIW_COCOSO vs others
results = []
base_model = "IDOCRIW_COCOSO"

for model in models:
    if model != base_model:
        stat, p_value = wilcoxon(errors[base_model], errors[model])
        base_median = np.median(errors[base_model])
        comp_median = np.median(errors[model])
        results.append({
            "Comparison": f"{base_model} vs {model}",
            "Wilcoxon_stat": stat,
            "p_value": p_value,
            "Base_median_error": base_median,
            "Comparison_median_error": comp_median
        })

# Convert to DataFrame and save
results_df = pd.DataFrame(results)
results_df.to_csv("wilcoxon_pairwise_results.csv", index=False)

print("Wilcoxon Pairwise Test Results:")
print(results_df)

# Boxplot of absolute errors
plt.figure(figsize=(8, 6))
plt.boxplot([errors[m] for m in models], labels=models)
plt.ylabel("Absolute Error")
plt.title("Absolute Errors of Models vs Storativity")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("model_errors_boxplot.tif", dpi=300)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import roc_auc_score
from scipy.stats import wilcoxon

# ==== 1. Load CSV ====
df = pd.read_csv("winconxon_test_data.csv")

# ==== 2. K-means binarization of Storativity ====
storativity_vals = df["Storativity"].values.reshape(-1, 1)
storativity_km = KMeans(n_clusters=2, random_state=42).fit(storativity_vals)
df["Storativity_Class"] = storativity_km.labels_

# Ensure 1 = high storativity, 0 = low
if df.groupby("Storativity_Class")["Storativity"].mean()[0] > df.groupby("Storativity_Class")["Storativity"].mean()[1]:
    df["Storativity_Class"] = 1 - df["Storativity_Class"]

# ==== 3. Model list ====
models = ["IDOCRIW_COCOSO", "CRITIC_COCOSO", "MeanWeighted", "MeanWeighted_COCOSO"]

# ==== 4. AUC calculation ====
auc_results = {}
for model in models:
    auc_val = roc_auc_score(df["Storativity_Class"], df[model])
    auc_results[model] = auc_val

# ==== 5. Wilcoxon signed-rank test ====
wilcoxon_results = {}
ido_model = "IDOCRIW_COCOSO"
for model in models:
    if model != ido_model:
        stat, p_value = wilcoxon(df[ido_model], df[model])
        wilcoxon_results[model] = (stat, p_value)

# ==== 6. Create results table ====
results_table = []
for model in models:
    if model == ido_model:
        results_table.append([model, auc_results[model], "-", "-", "-"])
    else:
        stat, p_val = wilcoxon_results[model]
        significance = "Yes" if p_val < 0.05 else "No"
        results_table.append([model, auc_results[model], stat, p_val, significance])

results_df = pd.DataFrame(results_table,
                          columns=["Model", "AUC", "Wilcoxon Stat", "p-value", "Significant (p<0.05)"])

# ==== 7. Save table ====
results_df.to_csv("model_validation_results_winconxon.csv", index=False)

print("\n=== Validation Results ===")
print(results_df)
print("\nTable saved as model_validation_results.csv")
```

```python
# Bootsrapping for the model's performance
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv("winconxon_test_data_boostrapping.csv")

# Model names
models = ["IDOCRIW-COCOSO", "CRITIC-COCOSO", "Mean Weighted COCOSO"]

# Create binary ground truth from storativity (median split)
threshold = np.median(df["Storativity"])
y_true = (df["Storativity"] >= threshold).astype(int).values

# Bootstrap settings
n_iterations = 1000
n_size = len(df)
rng = np.random.default_rng(42)

results_summary = {}

# Run bootstrapping for each model
for model in models:
    y_pred = df[model].values
    boot_aucs = []
    for i in range(n_iterations):
        sample_idx = rng.choice(np.arange(n_size), size=n_size, replace=True)
        auc = roc_auc_score(y_true[sample_idx], y_pred[sample_idx])
        boot_aucs.append(auc)

    boot_aucs = np.array(boot_aucs)
    results_summary[model] = {
        "Mean_AUC": np.mean(boot_aucs),
        "Std_Dev": np.std(boot_aucs),
        "CI_95_Lower": np.percentile(boot_aucs, 2.5),
        "CI_95_Upper": np.percentile(boot_aucs, 97.5),
        "All_AUCs": boot_aucs
    }

# Display results
print("=== Bootstrapping AUC Results ===")
for model, stats in results_summary.items():
    print(f"{model}: Mean={stats['Mean_AUC']:.3f}, Std={stats['Std_Dev']:.3f}, "
          f"95% CI=({stats['CI_95_Lower']:.3f}, {stats['CI_95_Upper']:.3f})")

# Plot histogram comparison
plt.figure(figsize=(10, 6))
for model, stats in results_summary.items():
    plt.hist(stats["All_AUCs"], bins=20, alpha=0.5, label=model)
plt.xlabel("AUC Score")
plt.ylabel("Frequency")
plt.title("Bootstrapped AUC Distributions (1000 Iterations)")
plt.legend()
plt.tight_layout()
plt.savefig("bootstrap_auc_comparison.tif", dpi=1000, format='tiff')
plt.show()
```

```python
# mann-whitney test 

import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

# Assuming you have bootstrapped AUC values stored in a dictionary
# Example: bootstrapped results from previous code
np.random.seed(42)  # For reproducibility
n_bootstraps = 1000

# Replace these with your actual bootstrapped values
auc_idocriw = np.random.normal(0.851, 0.027, n_bootstraps)
auc_critic = np.random.normal(0.701, 0.038, n_bootstraps)
auc_meanw = np.random.normal(0.725, 0.037, n_bootstraps)
auc_meanwcocoso = np.random.normal(0.770, 0.037, n_bootstraps)

models = {
    "CRITIC_COCOSO": auc_critic,
    "Mean Weighted Model": auc_meanw,
    "Mean Weighted COCOSO": auc_meanwcocoso
}

print("=== MannWhitney U Test Results ===")
for name, auc_values in models.items():
    stat, p = mannwhitneyu(auc_idocriw, auc_values, alternative="greater")
    print(f"IDOCRIW_COCOSO vs {name}: U-stat = {stat:.2f}, p-value = {p:.5f}")
    if p < 0.05:
        print("   Statistically significant superiority of IDOCRIWCOCOSO")
    else:
        print("   No significant difference")
```
